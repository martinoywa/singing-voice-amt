# Results Interpretation Figures

```python
# -----------------------------------------------------------
# Performance Analysis of Timbre-Transferred Audio
# -----------------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from math import pi

# Configure plotting aesthetics
sns.set(style="whitegrid", context="talk", font_scale=0.9)

# -----------------------------------------------------------
# 1. Load all model results
# -----------------------------------------------------------
models = ["MT3", "basic_pitch"]
instruments = ["flute", "trumpet", "tenor_saxophone", "violin"]

all_results = []

for model in models:
    for inst in instruments:
        fname = f"evaluation_results_{model}_{inst}.csv"
        if os.path.exists(fname):
            df = pd.read_csv(fname)
            df["Model"] = model
            df["Instrument"] = inst.capitalize().replace("_", " ")
            all_results.append(df)

# Combine all into one DataFrame
results = pd.concat(all_results, ignore_index=True)

# -----------------------------------------------------------
# 2. Compute average metrics per instrument & model
# -----------------------------------------------------------
metrics = ["Precision", "Recall", "F1", "Onset Precision", "Onset Recall", "Onset F1", "Onset-Offset F1", "TER"]
avg_results = results.groupby(["Model", "Instrument"])[metrics].mean().reset_index()

# Save the table for reference
avg_results.to_csv("transcription_summary.csv", index=False)
print(avg_results)

# -----------------------------------------------------------
# 3. Bar Plots: F1 and Onset F1 across instruments
# -----------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_results, x="Instrument", y="F1", hue="Model")
plt.title("F1-Score Across Timbre-Transferred Instruments")
plt.ylabel("Average F1-Score")
plt.xlabel("Instrument")
plt.legend(title="Model")
plt.tight_layout()
plt.savefig("F1_across_instruments.png", dpi=300)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_results, x="Instrument", y="Onset F1", hue="Model")
plt.title("Onset F1 Across Timbre-Transferred Instruments")
plt.ylabel("Average Onset F1")
plt.xlabel("Instrument")
plt.legend(title="Model")
plt.tight_layout()
plt.savefig("Onset_F1_across_instruments.png", dpi=300)
plt.show()

# -----------------------------------------------------------
# 4. Violin Plots: Distribution of F1 and Onset F1
# -----------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(data=results, x="Instrument", y="F1", hue="Model", palette="coolwarm")
plt.title("Distribution of F1-Scores Across Timbre-Transferred Audio")
plt.ylabel("F1-Score")
plt.xlabel("Instrument")
plt.tight_layout()
plt.savefig("F1_distribution_violin.png", dpi=300)
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(data=results, x="Instrument", y="Onset F1", hue="Model", split=True, inner="quart")
plt.title("Distribution of Onset F1 Across Timbre-Transferred Audio")
plt.ylabel("Onset F1")
plt.xlabel("Instrument")
plt.tight_layout()
plt.savefig("Onset_F1_distribution_violin.png", dpi=300)
plt.show()

# -----------------------------------------------------------
# 5. Heatmap: Correlation among metrics (for each model)
# -----------------------------------------------------------
for model in results["Model"].unique():
    plt.figure(figsize=(8, 6))
    corr = results[results["Model"] == model][metrics].corr()
    sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
    plt.title(f"Metric Correlation Heatmap â€“ {model}")
    plt.tight_layout()
    plt.savefig(f"heatmap_correlation_{model}.png", dpi=300)
    plt.show()

# -----------------------------------------------------------
# 6. Radar Plot: Per-instrument performance profile
# -----------------------------------------------------------
def radar_plot(data, title):
    categories = list(data.columns[2:])  # Skip Model & Instrument
    N = len(categories)

    for i, row in data.iterrows():
        values = row[categories].values.flatten().tolist()
        values += values[:1]  # Close the circle
        angles = [n / float(N) * 2 * pi for n in range(N)]
        angles += angles[:1]

        plt.figure(figsize=(6, 6))
        ax = plt.subplot(111, polar=True)
        plt.xticks(angles[:-1], categories, color='grey', size=8)
        ax.plot(angles, values, linewidth=2, linestyle='solid', label=row["Model"])
        ax.fill(angles, values, alpha=0.3)
        plt.title(f"{row['Instrument']} ({row['Model']})", size=12, pad=20)
        plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
        plt.tight_layout()
        plt.savefig(f"radar_{row['Model']}_{row['Instrument'].replace(' ', '_')}.png", dpi=300)
        plt.close()

# Generate radar plots
radar_plot(avg_results, "Instrument-wise Performance Profiles")

# -----------------------------------------------------------
# 7. Optional: Table display formatting for publication
# -----------------------------------------------------------
formatted_table = avg_results.copy()
formatted_table[metrics] = formatted_table[metrics].applymap(lambda x: f"{x:.3f}")
print("\nFormatted Summary Table for Publication:\n")
print(formatted_table.to_markdown(index=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("transcription_summary.csv")

# The metric columns in the file
metrics = [
    'Precision',
    'Recall',
    'F1',
    'Onset Precision',
    'Onset Recall',
    'Onset F1',
    'Onset-Offset F1'
]

# Melt wide â†’ long
df_long = df.melt(
    id_vars=['Model', 'Instrument'],
    value_vars=metrics,
    var_name='Metric',
    value_name='Score'
)

# We allow zeros here (so basic_pitch still shows up); just drop NaN if any
df_long = df_long.dropna(subset=['Score', 'Model', 'Instrument'])

# Average score per (Model, Instrument, Metric)
summary = (
    df_long
    .groupby(['Model', 'Instrument', 'Metric'], as_index=False)['Score']
    .mean()
)

# Order models explicitly: basic_pitch on top, MT3 on bottom
models_order = ['basic_pitch', 'MT3']

# Consistent instrument ordering in the legend
instrument_order = sorted(summary['Instrument'].unique())

sns.set_theme(style="whitegrid")

# One subplot per model, stacked vertically
fig, axes = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(12, 10),
    sharex=True,
    sharey=True
)

for ax, model in zip(axes, models_order):
    model_data = summary[summary['Model'] == model]

    # Plot bar chart: x = Metric, grouped bars = Instrument
    sns.barplot(
        data=model_data,
        x='Metric',
        y='Score',
        hue='Instrument',
        hue_order=instrument_order,
        errorbar=None,               # seaborn 0.11: disables CI bars
        ax=ax
    )

    ax.set_title(f"{model.title() if model == 'basic_pitch' else model} Mean Performance Across Instruments", fontsize=14)
    ax.set_ylabel("Score", fontsize=12)
    ax.set_ylim(0, 1)          # compare systems on same scale
    ax.set_xlabel("")          # we'll label x only on bottom plot
    ax.legend(title="Instrument", loc="upper right")

# Bottom subplot gets the x-axis label
axes[-1].set_xlabel("Metric", fontsize=12)

plt.tight_layout()
plt.show()
```
